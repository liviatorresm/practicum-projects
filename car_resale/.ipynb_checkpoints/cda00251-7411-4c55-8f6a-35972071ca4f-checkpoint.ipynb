## Lidando com valores ausentes

### Variável model_year

```python
df_car['model_year'].isna().sum()
```

```python
df_car['model_year'].isna().sum()/df_car.shape[0]*100
```

Por se tratar de 7% dos dados, irei substituir esses valores pela mediana.

```python
conditionlist = [
    (df_car['model_year'].isna() == True) ]
choicelist = [df_car['model_year'].median()]
try:
    df_car['model_year'] = np.select(conditionlist, choicelist, default=df_car['model_year'].repeat(1))
except Exception as e:
    print(e)
```

```python
#verificando se ainda há valores ausentes na variável
df_car['model_year'].isna().sum()
```

### Variável quilometragem

```python
# Verificando se há relação entre o ano do veículo e a quilometragem
df_car['model_year'].corr(df_car['odometer'])
```

Há uma correlação negativa moderada entre o ano do modelo do veículo e quilometragem. Irei preencher os valores ausentes com a mediana de quilometragem por ano do modelo do veículo.

```python
#Separanda a variável model_year em quartis
pd.qcut(df_car['model_year'],4)
```

```python
#Criando categorias com base nos quartis de ano do modelo do veículo
def categoria_model_year(row):
    if row <= 2006:
        return 'até_2006'
    elif row <= 2011:
        return 'até_2011'
    elif row <= 2014:
        return 'até_2014'
    else:
        return 'até_2019'
```

```python
#testando a função
try:
    df_car['model_year'].apply(categoria_model_year)
except Exception as e:
    print(e)
```

```python
#aplicando a função e atribuindo a uma nova coluna no dataframe
df_car['model_year_cat'] = df_car['model_year'].apply(categoria_model_year)
```

```python
#descobrindo a mediana de quilometragem pelos quartis de ano do modelo
df_car.pivot_table(index='model_year_cat', values='odometer', aggfunc='median')
```

```python
#Testando e preenchendo os valores ausentes de quilometragem com a mediana por quartis
conditionlist = [
    (df_car['model_year_cat'] == 'até_2006') & (df_car['odometer'].isna() == True) ,
    (df_car['model_year_cat'] == 'até_2011') & (df_car['odometer'].isna() == True),
    (df_car['model_year_cat'] == 'até_2014') & (df_car['odometer'].isna() == True),
    (df_car['model_year_cat'] == 'até_2019') & (df_car['odometer'].isna() == True)]
choicelist = [58116.0, 132724.0, 99748.0, 58890.0]
try:
    df_car['odometer'] = np.select(conditionlist, choicelist, default=df_car['odometer'].repeat(1))
except Exception as e:
    print(e)
```

```python
#verificando se ainda há valores ausentes na variável
df_car['odometer'].isna().sum()
```

### Variável cilindros

```python
#Percentual de valores ausentes em relação aos dados
df_car['cylinders'].isna().sum()/df_car.shape[0]*100
```

```python
#Irei separar por mediana de cilindros pelo tipo de carros, para preencher os valores ausentes
df_car.pivot_table(index='type', values='cylinders', aggfunc='median')
```

```python
#Testando e preenchendo os valores ausentes de cilindros com a mediana por tipo de veículo
conditionlist = [
    (df_car['type'] == 'bus') & (df_car['cylinders'].isna() == True) ,
    (df_car['type'].isin(['convertible','pickup', 'truck'])) & (df_car['cylinders'].isna() == True),
    (df_car['type'].isin(['hatchback', 'sedan', 'wagon'])) & (df_car['cylinders'].isna() == True),
    (df_car['type'].isin(['SUV', 'coupe', 'mini-van', 'offroad', 'other', 'van'])) & (df_car['cylinders'].isna() == True)
     ]
choicelist = [10.0, 8.0, 4.0, 6.0]
try:
    df_car['cylinders'] = np.select(conditionlist, choicelist, default=df_car['cylinders'].repeat(1))
except Exception as e:
    print(e)
```

```python
#verificando se ainda há valores ausentes na variável
df_car['cylinders'].isna().sum()
```

### Variável cor do veículo

```python
#Percentual de valores ausentes em relação aos dados
df_car['paint_color'].isna().sum()/df_car.shape[0]*100
```

São muitos carros em que não foram registrados as cores, por esse fato irei preencher os valores ausentes pela string 'unknown'

```python
df_car['paint_color'] = df_car['paint_color'].fillna('unknown')
```

```python
#verificando se ainda há valores ausentes na variável
df_car['paint_color'].isna().sum()
```

### Variável 4x4

Irei preencher os valores ausentes por 0 e então mudar o tipo de variável para booleana

```python
#Preenchendo os valores ausentes que correspondem a False
df_car['is_4wd']= df_car['is_4wd'].fillna(0)
#Mudando o tipo de variável
df_car['is_4wd'] = df_car['is_4wd'].astype('bool')
```

```python
#verificando se ainda há valores ausentes na variável
df_car['is_4wd'].isna().sum()
```

```python
#verificando tipo da variável
df_car['is_4wd'].dtype
```

## Correção dos tipos de dados

```python
#Verificando os tipos de dados
df_car.info()
```

```python
#Passando o preço para tipo float
df_car['price'] = df_car['price'].astype('float')
```

```python
#Passando o ano do modelo para tipo int
df_car['model_year'] = df_car['model_year'].astype('int')
```

```python
#Passando o data em que o anúcio foi registrado para tipo datetime
df_car['date_posted'] = pd.to_datetime(df_car['date_posted'], format='%Y-%m-%d')
```

```python
#Verificando os tipos das variáveis após correções
df_car.info()
```

```python
# Escrevendo função que retonará um valor numérico para a variável condição do veículo
def escala_condition(row):
    if row == 'new':
        return 5
    elif row == 'like new':
        return 4
    elif row == 'excellent':
        return 3
    elif row == 'good':
        return 2
    elif row == 'fair':
        return 1
    else:
        return 0
```

```python
#Testando a função
try:
    df_car['condition'] = df_car['condition'].apply(escala_condition)
except Exception as e:
    print(e)
```

```python
#Verificando resultado
df_car['condition'].unique()
```

## Lidando com outliers

```python
df_car.odometer.describe()
```

```python
# Determinando o limite inferior para valores atípicos
df_car['price_min'] = df_car['price'] < 1000.00
df_car['car_age_min'] = df_car['car_age'] < 3
df_car['odometer_min'] = df_car['odometer'] < 61030.000000
df_car['condition_min'] = df_car['condition'] < 1
```

```python
# Determinando o limite superior para valores atípicos
df_car['price_max'] = df_car['price'] > 50000.00
df_car['car_age_max'] = df_car['car_age'] > 40
df_car['odometer_max'] = df_car['odometer'] > 4000000.000000
df_car['condition_max'] = df_car['condition'] > 4
```

```python
#Criando novo dataframe sem valores atípicos de preço
data_filtro= df_car.query('price_min == False & price_max == False')
```

```python
#Filtrando o dataframe anterior sem valores atípicos de idade do veículo
data_filtro= data_filtro.query('car_age_min == False & car_age_max == False')
```

```python
#Filtrando o dataframe anterior sem valores atípicos de quilometragem
data_filtro= data_filtro.query('odometer_min == False & odometer_max == False')
```

```python
#Filtrando o dataframe anterior sem valores atípicos de condition
data_filtro= data_filtro.query('condition_min == False & condition_max == False')
```

```python
data_filtro.shape[0]/df_car.shape[0]
```

Restaram 71% dos dados.

## Mediana de preços por cada tipo de veículo

```python
#armazenando contagem dos anúncios em uma variável
count_anuncio_type = data_filtro.pivot_table(index= 'type', values='id', aggfunc='count')
count_anuncio_type
```

```python
#armazenando a mediana do preço dos veículos em uma variável
median_price_type = data_filtro.pivot_table(index= 'type', values='price', aggfunc='median')
median_price_type
```

```python
#concatenando dos dois dataframes 
type_price= pd.concat([count_anuncio_type, median_price_type], axis=1)
#renomeando as colunas
type_price.columns=['count','mean']
#ordenando por contagem em ordem descrescente 
type_price.sort_values(by='count', ascending=False, inplace=True)
```

```python
#passandoa o index com o nome dos tipos para uma coluna
type_price = type_price.reset_index()
```

```python
#verificando resultado
type_price
```

```python
#Gráfico em barra 
type_price.plot(x='type', y='count', kind='bar',  figsize=(12, 8), grid=True,fontsize=12)
plt.title('Quantidade de anúncios por tipo de veículo')
plt.show()
```

**Impressõe**
<div>SUV e truck são os tipos de veículos com maior número de anúncios</div>

É possível concluir que há uma correlação negativa moderada entre a idade do veículo e o preço. Os carros mais velhos parecem ter preços menores que os mais novo. Já em quilometragem e a condição do veículo visualizamos uma correlação negativa fraca. 

Há uma correlação positiva muito fraca entre o preço e o estado dos veículos, isto é quanto melhor o estado maior o preço. 

Parece não haver diferença importante entre o preço e o tipo de transmissão, pois os dados são mais espaçados e possui mediana menor.
Os carros com transmissão do tipo manual parecem ter valores de preço menor, porém outros tipos de transmissão tem preço mais elevado. Em relação a cor, os carros branco, amarelo e preto parecem ter preço maiores.
