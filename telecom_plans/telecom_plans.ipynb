#### 4.3.2 Calculando a receita mensal de minutos, mensagens e volumes de dados por mês, plano e usuário

```python
# Criando tabela dinâmica que retorna o contagem de chamadas utilizados por usuários, mês e plano
n_calls = df_calls.pivot_table(index=['user_id','plan', 'month', 'age', 'reg_date','city'], values='id_call', aggfunc='count')
# Verificando resultado
n_calls
```

```python
# Criando tabela dinâmica que retorna o somatório de minutos utilizados por usuários, mês e plano
n_minutes = df_calls.pivot_table(index=['user_id','plan', 'month', 'age', 'reg_date', 'city'], values='duration', aggfunc='sum')
# Verificando resultado
n_minutes
```

```python
# Criando tabela dinâmica que retorna o somatório de mb utilizados por usuários, mês e plano
volume_dados = df_internet.pivot_table(index=['user_id','plan', 'month', 'age', 'reg_date', 'city'], values='mb_used', aggfunc='sum')
# Verificando resultado
volume_dados
```

```python
# Como a receita é calculada com base no valor do Gigabyte excedido, criei uma função que retorna a quantidade de Gb excedida pelo usuário
# Essa função compara o nome do plano do user_id e compara com o nome da tabela df_plans
# Cada valor de mb utilizado por mês e subtrai do valor de mb incluído de acordo com o plano. Isso é o mb excedido
# Em seguida esse valor é dividido por 1024 com intuito de converte-lo em Gb
# Após converter para Gb a função retorna o valor caso ele seja positivo, caso contrário ele retorna 0, que significa que o 
# usuário não ultrapassou a quantidade de mb incluído no plano
def calc_gb(row, col1, col2):
    plan_name = row.name[1]
    if plan_name == df_plans.loc[0, 'plan_name']:
        result = (row[col1]- df_plans.loc[0, col2]) / 1024
        if result > 0:
            return result
        else:
            return 0 
    elif plan_name == df_plans.loc[1, 'plan_name']: 
        result = (row[col1]- df_plans.loc[1, col2]) / 1024
        if result > 0:
            return result
        else:
            return 0
    else:
        return 'inválido'
```

```python
# Aplicando a função no dataframe e salvando em uma nova coluna 'gb_exceed'
volume_dados['gb_exceed'] = volume_dados.apply(calc_gb, axis=1, args=('mb_used', 'mb_per_month_included'))
```

```python
#Arredondando os valores
volume_dados['gb_exceed'] = volume_dados['gb_exceed'].round()
```

```python
# Verificando resultados
volume_dados
```

```python
# Criando tabela dinamica que retorna a contagem de mensagens enviadas por usuários, mês e plano
n_messages = df_messages.pivot_table(index=['user_id','plan', 'month', 'age', 'reg_date', 'city'], values='id_messages', aggfunc='count')
# Verificando resultados
n_messages
```

```python
# Criando função que retorna o cálculo da receita mensal de cada usuário por mês
# Essa função pega cada valor na coluna 1 (nome do plano) e armazena na variável plan_name
# Em seguida a função compara com o nome do plano na tabela df_plans que armazena as informações de cada plano
# Se for igual ao nome do plano presente na primeira linha (0) ou segunda (1) ele irá subtrair o que está incluído no plano do que foi utilizado pelo usuário,
# em seguida multiplica esse valor pelo valor que deve ser pago de quantidade excedida e soma com o valor do plano
# a função retorna a receita mensal
def calculate_revenue(row, col1, col2, col3,col4):
    plan_name = row.name[1]
    if plan_name == df_plans.loc[0, 'plan_name']:
        return (row[col1]- df_plans.loc[0, col2]) * df_plans.loc[0, col3] + df_plans.loc[0, col4]
    else:
        return (row[col1]- df_plans.loc[1, col2]) * df_plans.loc[1, col3] + df_plans.loc[1, col4]
    pass
```

```python
# Aplicando a função para calcular a receita mensal de minutos no dataframe e salvando em uma nova coluna 'receita_mensal'
n_minutes['receita_mensal'] = n_minutes.apply(calculate_revenue, axis=1, args=('duration', 'minutes_included', 'usd_per_minute', 'usd_monthly_pay'))
# Verificando resultados
n_minutes
```

```python

(n_minutes.apply(calculate_revenue, axis=1, args=('duration', 'minutes_included', 'usd_per_minute', 'usd_monthly_pay')) < 0).sum()
```

```python
# Aplicando a função para calcular a receita mensal de mensagens enviadas no dataframe e salvando em uma nova coluna 'receita_mensal'
n_messages['receita_mensal'] = n_messages.apply(calculate_revenue, axis=1, args=('id_messages', 'messages_included', 'usd_per_message', 'usd_monthly_pay'))
# Verificando resultados
n_messages
```

```python
# Como a coluna de volume de dados já possui a quantidade de mb excedido e convertida para Gb
# Essa função irá pegar o valor da coluna 'gb_exceed' e multiplicar pelo valor do gb excedido e somar com o valor mensal do plano,
# da mesma forma que a função calculate_revenue ele irá comparar os nomes dos plano para fazer o calculo de acordo com o plano contratado por cada usuário
def calculate_revenue_gb(row, col1, col2, col3):
    plan_name = row.name[1]
    if plan_name == df_plans.loc[0, 'plan_name']:
        return row[col1] * df_plans.loc[0, col2] + df_plans.loc[0, col3]
    else:
        return row[col1] * df_plans.loc[1, col2] + df_plans.loc[1, col3]
    pass
```

```python
#  Aplicando a função para calcular a receita mensal de volume de dados no dataframe e salvando em uma nova coluna 'receita_mensal'
volume_dados['receita_mensal'] = volume_dados.apply(calculate_revenue_gb, axis=1, args=('gb_exceed', 'usd_per_gb', 'usd_monthly_pay'))
# Verificando resultados
volume_dados
```

## 5. Testes de hipóteses

As hipóteses são: </p>
* A média da receita mensal dos usuários dos planos ultimate e surf são diferentes?</p>
    * Hipótese nula (H0): As receitas não são diferentes.</p>
    * Hipótese alternativa (Ha): As receitas são diferentes.</p>
* A receita média dos usuários da região de New York-Newark-Jersey City e das demais regiões são diferentes?</p>
    * Hipótese nula (H0): A receita dos usuários da região de New York-Newark-Jersey City e das demais regiões não são diferentes.</p>
    * Hipótese alternativa (Ha): A receita dos usuários da região de New York-Newark-Jersey City e das demais regiões são diferentes.</p>  
Para esse trabalho irei utilizar o valor de p 5%.

**Primeira hipótese: A receita média dos usuários dos planos Ultimate e Surf são diferentes.**

```python
# Armazenando os valores das receitas do grupo com plano surf na variável user_plan_surf
users_plan_surf = df_concat.query('plan=="surf"')['receita_total']
# Verificando resultado
users_plan_surf
```

```python
# Armazenando os valores das receitas do grupo com plano ultimate na variável user_plan_ultimate
users_plan_ultimate = df_concat.query('plan=="ultimate"')['receita_total']
# Verificando resultado
users_plan_ultimate
```

```python
# Armazenando o valor de alpha 5% na variável alpha
alpha = 0.05
# Chamando função do teste t para amostras independentes e armazenando na variável resultado
resultado = st.ttest_ind(users_plan_surf, users_plan_ultimate, equal_var=False)

print('p-value: %.10f' % resultado.pvalue)

if resultado.pvalue < alpha:
    print("Nós rejeitamos a hipótese nula.")
else:
    print("Nós não podemos rejeitar a hipótese nula.")
```

**Segunda hipótese: A receita média dos usuários da região de New York-Newark-Jersey City e das demais regiões são diferentes?**
